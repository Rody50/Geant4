## Week1, FCI

```python
#load the results
correlation_energy = pd.read_csv(resultDir + 'FCI/correlation_energy.txt', sep = "\s+", header = None)
g_value = correlation_energy[0]
corr_en = correlation_energy[1]

#set plot style 
sns.set(font_scale = 2)
sns.set_style('white')

#make plot 
fig, ax = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(15)
ax.plot(g_value, corr_en, marker = 'o', markersize = 10)
ax.plot([-1.1, 1.1],[0, 0], linestyle = '--', color = 'black')

plt.title(r'Correlation energy vs. Interaction strength', fontsize = '25')
plt.ylabel(r'$\Delta E$', fontsize = 20)
plt.xlabel(r'$g$', fontsize = 20)
plt.xlim(-1.1,1.1)
```

```python
#load the results
nSteps8 = pd.read_csv(resultDir + 'CCM/CCM_corr_en-steps_N_8_4_g_-1.000000.txt', sep = "\s+", header = None)
nSteps16 = pd.read_csv(resultDir + 'CCM/CCM_corr_en-steps_N_16_4_g_-1.000000.txt', sep = "\s+", header = None)
nSteps20 = pd.read_csv(resultDir + 'CCM/CCM_corr_en-steps_N_20_4_g_-1.000000.txt', sep = "\s+", header = None)
#correlation_energy = pd.read_csv(resultDir + 'CCM/CCM_N_8_A_4.txt', sep = "\s+", header = None)

corrSteps8 = nSteps8[0]
steps8 = range(len(corrSteps8))
corrSteps16 = nSteps16[0]
steps16 = range(len(corrSteps16))
corrSteps20 = nSteps20[0]
steps20 = range(len(corrSteps20))

#set plot style 
sns.set(font_scale = 2)
sns.set_style('white')

#make plot 
fig, ax = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(15)
ax.plot(steps8, corrSteps8, marker = 'o', markersize = 10, label = 'N = 8, A = 4')
ax.plot(steps16, corrSteps16, marker = 'o', markersize = 10, label = 'N = 16, A = 4')
ax.plot(steps20, corrSteps20, marker = 'o', markersize = 10, label = 'N = 20, A = 4')
#ax.plot([-1.1, 1.1],[0, 0], linestyle = '--', color = 'black')

plt.title(r'Convergence', fontsize = 25)
plt.ylabel(r'$\Delta E$', fontsize = 20)
plt.xlabel(r'$steps$', fontsize = 20)
plt.legend(loc = 'lower right')
plt.xlim(-1,len(corrSteps20))
```
