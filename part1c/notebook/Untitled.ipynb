```python
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
correlation_energy = pd.read_csv('~/Documents/TUD/TALENT2018/Geant4/part1c/results/correlation_energy.txt', sep = "\s+", header = None)

g_value = correlation_energy[0]
corr_en = correlation_energy[1]
sns.set(font_scale = 2)

fig, ax = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(15)
ax.plot(g_value, corr_en, marker = 'o', linestyle = 'None', markersize = 10, label = r'$G^e_p$')

plt.title(r'Nucleon Formfactors', fontsize = '25')
plt.ylabel(r'$\Delta E$', fontsize = 20)
plt.xlabel(r'$g$', fontsize = 20)
plt.legend()
plt.xlim(-1,1)
```
